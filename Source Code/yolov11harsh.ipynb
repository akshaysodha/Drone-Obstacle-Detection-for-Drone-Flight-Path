```python
from ultralytics import YOLO
# Load the YOLOv8 model (pretrained weights)
model = YOLO('/kaggle/input/harsh1/final-year-test/yolo11n.pt')  # Use 'yolov8s.pt' for a larger model if needed

# Train the model
model.train(
    data="/kaggle/input/datayaml/dataset (5).yaml",  # Path to dataset YAML
    epochs=500,  # Number of training epochs
    imgsz=640,  # Image size (can be adjusted)
    batch=16,  # Batch size
    workers=4,  # Number of workers
    project="/kaggle/working/final-year-test+",  # Save directory
    name="yolovnew_train",  # Experiment name
    exist_ok=True,  # Overwrite existing experiment
    pretrained=True,  # Use pretrained weights
    optimizer="SGD"  # Use SGD optimizer (default is AdamW)
)
```
